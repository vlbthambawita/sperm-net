```python
fattyacid_serum_csv = "/global/D1/homes/vajira/data/sperm/visem/visem-dataset/fatty_acids_serum.csv"
fattyacid_spermatoza_csv = "/global/D1/homes/vajira/data/sperm/visem/visem-dataset/fatty_acids_spermatoza.csv"
participant_related_csv = "/global/D1/homes/vajira/data/sperm/visem/visem-dataset/participant_related_data.csv"
semen_analysis_data = "/global/D1/homes/vajira/data/sperm/visem/visem-dataset/semen_analysis_data.csv"
sex_hormones_csv = "/global/D1/homes/vajira/data/sperm/visem/visem-dataset/sex_hormones.csv"

```

```python
# Read all csv files into dataframes
import pandas as pd

fattyacid_serum_df = pd.read_csv(fattyacid_serum_csv, delimiter=';')
fattyacid_spermatoza_df = pd.read_csv(fattyacid_spermatoza_csv, delimiter=';')
participant_related_df = pd.read_csv(participant_related_csv, delimiter=';')
semen_analysis_data_df = pd.read_csv(semen_analysis_data, delimiter=';')
sex_hormones_df = pd.read_csv(sex_hormones_csv, delimiter=';')
```

```python
# combine all dataframes into one using ID as the key
df = pd.merge(fattyacid_serum_df, fattyacid_spermatoza_df, on='ID')
df = pd.merge(df, participant_related_df, on='ID')
df = pd.merge(df, semen_analysis_data_df, on='ID')
df = pd.merge(df, sex_hormones_df, on='ID')
```

```python
df.head()
```

```python
# change the ID column name to file_name
df.rename(columns={'ID': 'file_name'}, inplace=True)
```

```python
# add .avi to the file_name
df['file_name'] = df['file_name'].astype(str) + '.avi'
```

```python
# save the dataframe to a csv file named metadata.csv in the directory "/global/D1/homes/vajira/data/sperm/visem/visem-dataset"
df.to_csv('/global/D1/homes/vajira/data/sperm/visem/visem-dataset/metadata.csv', index=False)
```
